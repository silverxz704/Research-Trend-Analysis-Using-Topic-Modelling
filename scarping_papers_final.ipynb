```python

import xml.etree.ElementTree as ET
import requests
import urllib.request as libreq
import pandas as pd

 

with libreq.urlopen('http://export.arxiv.org/api/query?search_query=all:Economics&start=0&max_results=200') as url:
    xml_data = url.read()

# Parse the XML data

root = ET.fromstring(xml_data)

# Find and download all PDF links

pdf_links = []
pdf_name=[]
authors = []
published_dates = []
updated_dates = []
summaries = []
titles=[]

for entry in root.findall('{http://www.w3.org/2005/Atom}entry'):
    
    pdf_element = entry.find('{http://www.w3.org/2005/Atom}link[@title="pdf"]')
    
    title = entry.find('{http://www.w3.org/2005/Atom}title').text

    author = entry.find('{http://www.w3.org/2005/Atom}author/{http://www.w3.org/2005/Atom}name').text

    published_date = entry.find('{http://www.w3.org/2005/Atom}published').text

    updated_date = entry.find('{http://www.w3.org/2005/Atom}updated').text

    summary = entry.find('{http://www.w3.org/2005/Atom}summary').text

    if pdf_element is not None:

        pdf_links.append(pdf_element.get('href'))

       

    titles.append(title)

    authors.append(author)

    published_dates.append(published_date)

    updated_dates.append(updated_date)

    summaries.append(summary)

 

for idx, pdf_link in enumerate(pdf_links, start=1):

    response = requests.get(pdf_link)

    if response.status_code == 200:

        with open(f'paper_eco_{idx}.pdf', 'wb') as pdf_file:

            pdf_file.write(response.content)

        
        pdf_name.append(f"paper_eco_{idx}.pdf")

        print(f"PDF {idx} downloaded successfully.")

    else:

        print(f"Failed to download PDF {idx}.")

 

if not pdf_links:

    print("No PDF links found in the XML.")

   

data = {

    'title': titles,
    'author': authors,
    'published': published_dates,
    'updated': updated_dates,
    'summary': summaries,
    'file_name':pdf_name,
    'pdf_link': pdf_links
}

df = pd.DataFrame(data)
df.to_csv('info_eco.csv', index=False)
```

```python
from PyPDF2 import PdfReader
import pandas as pd



page_no=[]
page_data=[]
file_name=[]
for i in range(200):
    try:
        reader = PdfReader(f"paper_eco_{i+1}.pdf")
        for j in range(len(reader.pages)):
            file_name.append(f"paper_eco_{i+1}.pdf")
            page_no.append(j+1)
            page = reader.pages[j]
            parts = []
            def visitor_body(text, cm, tm, fontDict, fontSize):
                y = tm[5]
                if y > 12.5 and y < 770:
                    parts.append(text)
                    
            page.extract_text(visitor_text=visitor_body)
            text_body = "".join(parts)
            page_data.append(text_body)
        print(f"Data appended for paper_eco_{i+1}.pdf")
    except:
         continue

data = {'file_name': file_name,
        'page_number': page_no,
        'page_data': page_data}     
df2 = pd.DataFrame(data)
df = pd.read_csv('info_eco.csv')
merged_df = pd.merge(df2, df, on='file_name', how='left')
result_df = merged_df[['file_name', 'page_number', 'page_data', 'title', 'summary','published']]
result_df.to_csv('info_master_eco.csv', index=False, escapechar='\\')
```
